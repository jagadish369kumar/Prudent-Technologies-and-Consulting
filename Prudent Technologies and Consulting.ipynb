#### (C) Data Preprocessing

```python
print(data.isnull().sum()) # for checking missing values

data = data.dropna() # Drops row with missing values

data = pd.get_dummies(data, drop_first=True) # converting categorical var to num for machine understanding

data.head() # it displays first few rows of the pre processed data
```

### (D) Data Visualizations Using Libraries 

```python
import matplotlib.pyplot as plt # data visualization tool
import seaborn as sns, sklearn # data visualization tool
```

```python
data = pd.read_csv(r"C:\Users\JAGADISH\Downloads\archive (1)\customer_churn_data.csv")
data
```

### (1) Customer Demographics :

```python
# Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=data, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Senior Citizen status
plt.figure(figsize=(6, 4))
sns.countplot(x='SeniorCitizen', data=data, palette='Set2')
plt.title('Senior Citizen Distribution')
plt.xlabel('Senior Citizen')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Partner Status
plt.figure(figsize=(6, 4))
sns.countplot(x='Partner', data=data, palette='Set2')
plt.title('Partner Status')
plt.xlabel('Partner')
plt.ylabel('Count')
plt.show()

# Dependents Status
plt.figure(figsize=(6, 4))
sns.countplot(x='Dependents', data=data, palette='Set2')
plt.title('Dependents Status')
plt.xlabel('Dependents')
plt.ylabel('Count')
plt.show()
```

### (2) Usage Patterns :

```python
# Phone Service
plt.figure(figsize=(6, 4))
sns.countplot(x='PhoneService', data=data, palette='Set2')
plt.title('Phone Service')
plt.xlabel('Phone Service')
plt.ylabel('Count')
plt.show()

# Multiple Lines
plt.figure(figsize=(6, 4))
sns.countplot(x='MultipleLines', data=data, palette='Set2')
plt.title('Multiple Lines')
plt.xlabel('Multiple Lines')
plt.ylabel('Count')
plt.show()
```

```python
# Internet Service
plt.figure(figsize=(6, 4))
sns.countplot(x='InternetService', data=data, palette='Set2')
plt.title('Internet Service')
plt.xlabel('Internet Service')
plt.ylabel('Count')
plt.show()
```

```python
# Online Services
plt.figure(figsize=(10, 6))
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for service in services:
    plt.figure()
    sns.countplot(x=service, data=data, palette='Set2')
    plt.title(f'{service} Status')
    plt.xlabel(service)
    plt.ylabel('Count')
    plt.show()
```

### (3) Billing History :

```python
# Monthly Charges Distribution
plt.figure(figsize=(6, 4))
sns.histplot(data['MonthlyCharges'], bins=30, kde=True, color='blue')
plt.title('Distribution of Monthly Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
# Total Charges Distribution
plt.figure(figsize=(6, 4))
sns.histplot(data['TotalCharges'], bins=30, kde=True, color='green')
plt.title('Distribution of Total Charges')
plt.xlabel('Total Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
# Total Charges vs. Monthly Charges
plt.figure(figsize=(8, 4))
sns.scatterplot(x='MonthlyCharges', y='TotalCharges', hue='Churn', data=data, palette='Set1')
plt.title('Total Charges vs. Monthly Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.show()
```

### (4) Customer service interactions :

```python
# Online Services Interactions
plt.figure(figsize=(10, 6))
for service in services:
    plt.figure()
    sns.countplot(x=service, hue='Churn', data=data, palette='Set1')
    plt.title(f'Churn by {service} Status')
    plt.xlabel(service)
    plt.ylabel('Count')
    plt.legend(title='Churn')
    plt.show()
```
